```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv('Datasets/World cities population.csv', delimiter=',')
```

```python
dataset
```

```python
China_Cities = []
Cities_Population = []
```

```python
counter = 0
for i in range (0,41000):
    if(dataset['country'].iloc[i] == "China"):
        China_Cities.append(dataset['city'].iloc[i])
        Cities_Population.append(dataset['population'].iloc[i])
```

```python
China_Cities
```

```python
Cities_Population
```

```python
highest_population = []
cities_ = []
```

```python
while(len(cities_) < 5):
    highest = Cities_Population[0]
    counter = 0
    for x in range(0,len(China_Cities)):
        if(highest >= Cities_Population[x]):
            pass
        else:
            highest = Cities_Population[x]
            counter = x
    highest_population.append(highest)
    cities_.append(China_Cities[counter])
    del Cities_Population[counter]
    del China_Cities[counter]
```

```python
highest_population
```

```python
cities_
```

```python
plt.figure(figsize = (15,10))

plt.bar(cities_, highest_population, width= 0.9, align='center', edgecolor = 'black')
# This is the location for the annotated text
i = 1.0
j = 2000
# Annotating the bar plot with the values (total death count)
for i in range(len(cities_)):
    plt.annotate(highest_population[i], (-0.1 + i, highest_population[i] + j))
# Creating the legend of the bars in the plot
plt.legend(labels = ['Population'], fontsize = 12)
# Giving the tilte for the plot
plt.title("5 Cities in China That Has Highest Population", fontsize = 20, family = 'serif')
# Namimg the x and y axis
plt.xlabel('Cities', fontsize = 14, family = 'serif')
plt.ylabel('Population', fontsize = 14, family = 'serif')

plt.show()
```

```python
# Creating explode data
explode = (0.1, 0.1, 0.1)
 
# Creating color parameters
colors = ( "Green", "cyan", "Yellow")
 
# Wedge properties
wp = { 'linewidth' : 1.5, 'edgecolor' : "green" }
 
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\nPopulation:{:d}".format(pct, absolute)
 
# Creating plot
fig, ax = plt.subplots(figsize =(15, 7))
wedges, texts, autotexts = ax.pie(highest_,
                                  autopct = lambda pct: func(pct, highest_),
                                  explode = explode,
                                  labels = cities__,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="black"))
 
# Adding legend
legend = plt.legend(cities__,bbox_to_anchor =(1, 0, 0.5, 1), fontsize = 12)
legend.set_title("Legend", prop = {'size':15})
plt.setp(autotexts, size = 10, weight ="bold", family = "serif")
ax.set_title("3 Cities In The Philippines That Highest Population", fontsize = 18, family = "serif")
plt.show()
```
